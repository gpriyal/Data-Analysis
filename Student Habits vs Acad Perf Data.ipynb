# 3. Data Visualisation

```python
cat_col = df.select_dtypes(include='object').columns
cat_col = [col for col in cat_col]
cat_col.remove('student_id')
cat_col
```

```python
num_col = df.select_dtypes(exclude='object').columns
num_col = [col for col in num_col]
num_col
```

```python
plt.figure(figsize=(12,6))
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.hist(df[cat_col[i]])
    plt.title(f"Distribution of {cat_col[i]}")
plt.tight_layout()
```

```python
plt.figure(figsize=(12,10))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.hist(df[num_col[i]], bins=20, edgecolor='black')
    plt.title(f"Distribution of {num_col[i]}")
plt.tight_layout()
```

```python
sns.scatterplot(x='study_hours_per_day', y='exam_score', data=df)
plt.title("Study Hours v/s Exam Score")
plt.xlabel("Study Hours Per Day")
plt.ylabel("Exam Score")
```

```python
cols = ['social_media_hours', 'netflix_hours', 'attendance_percentage', 'sleep_hours']
plt.figure(figsize=(12,12))
for i in range(4):
    plt.subplot(2,2,i+1)
    sns.regplot(x=cols[i], y='exam_score', data=df)
    plt.title(f"{cols[i]} v/s Exam Score")
    plt.xlabel(cols[i])
    plt.ylabel("Exam Score")
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Between Student Habits and Exam Score')
```

```python
sns.barplot(data=df, x='internet_quality', y='exam_score', hue='mental_health_rating')
plt.title('Internet Quality vs Exam Score by Mental Health')
```

```python
plt.figure(figsize=(12,6))
for i in range(6):
    plt.subplot(2,3,i+1)
    sns.boxplot(x=df[cat_col[i]], y='exam_score', data=df)
plt.tight_layout()
```

```python
sns.lineplot(x='mental_health_rating', y='exam_score', data=df)
```
